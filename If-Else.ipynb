```python
"""WAP to accept the percentage from the user and display the grade according to the following criteria
Marks            Grade
>90               A
>80 and <=90      B
>60 and <=80      C
Below 60          D"""
```

```python
marks = float(input("Enter the Percentage of Student to get the Grade: "))
if marks > 100 or marks < 0:
    print("Invalid input: Marks should be between 0 and 100")
elif marks >= 90:
    print("Student gets the A Grade")
elif marks >= 80:
    print("Student gets the B Grade")
elif marks >= 60:
    print("Student gets the C Grade")
else:
    print("Student gets the D Grade")
```

```python
#Find the median of a list
list=[1,2,3,4,5,6,7,8]
l=len(list)
list.sort()
if l%2==0:
    median=(list[l//2]+list[l//2-1])/2
else:    
    median=list[l//2]
print("The median of the list is:",median)
```

```python
#Largest of Three Numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Check which number is greatest
if a >= b and a >= c:
    print(f"{a} is largest")
elif b >= a and b >= c:
    print(f"{b} is largest")
else:
    print(f"{c} is largest")
```

```python
#Check whether the year is leap year or not
year = int(input("Enter year: "))

# Leap year rule:
# Divisible by 4 AND not divisible by 100
# OR divisible by 400

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
#Password Strength Checker
password = input("Enter password: ")

if len(password) < 8:
    print("Weak Password - Too short")
elif password.isdigit() or password.isalpha():
    print("Weak Password - Use mix of letters and numbers")
elif not any(char.isupper() for char in password):
    print("Weak Password - Add at least one uppercase letter")
else:
    print("Strong Password")
```

```python
#ATM Machine Simulation
# ATM Machine Simulation
balance = 50000
pin = "1234"

entered_pin = input("Enter PIN: ")

if entered_pin == pin:
    print("1. Withdraw")
    print("2. Deposit")
    choice = input("Choose option: ")

    if choice == "1":
        amount = float(input("Enter withdrawal amount: "))
        if amount <= balance:
            balance -= amount
            print("Withdrawal Successful")
            print("Remaining Balance:", balance)
        else:
            print("Insufficient Balance")

    elif choice == "2":
        amount = float(input("Enter deposit amount: "))
        balance += amount
        print("Deposit Successful")
        print("Remaining Balance:", balance)

    else:
        print("Invalid Option")
else:
    print("Incorrect PIN")
```

```python
#Smart Traffic Fine System
speed = int(input("Enter vehicle speed: "))
zone = input("Enter zone (city/highway/school): ")

if zone == "city":
    limit = 60
elif zone == "highway":
    limit = 100
elif zone == "school":
    limit = 40
else:
    print("Invalid zone")
    limit = 0

if speed > limit:
    fine = (speed - limit) * 100
    print("Speed Limit Exceeded")
    print("Fine Amount:", fine)
else:
    print("Within Speed Limit")
```
