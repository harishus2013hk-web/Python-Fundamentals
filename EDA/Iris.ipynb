Calculating aggregate functions 

```python
m1=iris["sepal_length"].mean()
print(m1)

m2=iris["sepal_width"].median()
print(m2)

m3=iris["petal_length"].mode()
print(m3)

s=iris["petal_width"].sum()
print(s)

t=iris["species"].value_counts()
print(t)
```

Data Transformation

```python
iris["sepal_ratio"]=iris["sepal_length"]/iris["sepal_width"]        #New column
iris["sepal_ratio"]
iris.rename(columns={"sepal_ratio":"sepalratio"})
```

```python
iris["petal_ratio"]=iris["petal_length"]/iris["petal_width"]
iris
```

```python
iris["petal_area"]=iris["petal_length"]*iris["petal_width"]
iris["sepal_area"]=iris["sepal_length"]*iris["sepal_width"]
```

```python
iris["flower_size"]=iris["petal_area"]+iris["sepal_area"]
iris.head()
```

```python
iris.groupby("species").agg({"flower_size":["min","max","mean"]})
```

```python
iris["flower_size_group"]=pd.cut(iris["flower_size"],bins=[0,15,30,50],labels=["small","medium","large"])
iris.head()
```

```python
pd.pivot_table(data=iris,values=["sepal_length","sepal_width","petal_length","petal_width"],index="species",aggfunc="mean")
```

```python
plt.figure(figsize=(4,3))
a=sns.countplot(data=iris,x="species")
plt.bar_label(a.containers[0])
plt.show()
```

```python
a=sns.countplot(data=iris,x="flower_size_group",hue="species")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()
```
