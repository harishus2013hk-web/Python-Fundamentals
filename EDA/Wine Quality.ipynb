```python
sns.countplot(data=wine, x="quality")
plt.title("Distribution of Wine Quality")
plt.show()
```

```python
sns.histplot(data=wine, x="alcohol", kde=True)
plt.title("Alcohol Distribution")
plt.show()
```

```python
sns.histplot(data=wine, x="pH", kde=True)
plt.title("pH Distribution")
plt.show()
```

```python
sns.boxplot(data=wine, x="quality", y="alcohol")
plt.title("Alcohol vs Quality")
plt.show()
```

```python
sns.boxplot(data=wine, x="quality", y="volatile acidity")
plt.title("Volatile Acidity vs Quality")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(wine.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(wine,vars=["alcohol", "volatile acidity", "sulphates", "quality"],hue="quality")
plt.show()
```

```python
sns.kdeplot(data=wine, x="alcohol", hue="quality", fill=True)
plt.title("Alcohol Distribution by Wine Quality")
plt.show()
```

```python
sns.violinplot(data=wine, x="quality", y="alcohol")
plt.title("Alcohol vs Quality (Violin Plot)")
plt.show()
```

```python
sns.regplot(
    data=wine,
    x="alcohol",
    y="quality"
)
plt.title("Regression: Alcohol vs Quality")
plt.show()
```

```python
sns.scatterplot(
    data=wine,
    x="volatile acidity",
    y="alcohol",
    hue="quality"
)
plt.title("Volatile Acidity vs Alcohol")
plt.show()
```
