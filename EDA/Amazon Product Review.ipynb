```python
#Mean Rating per product
amazon.groupby("id")["reviews.rating"].mean().sort_values(ascending=False)
```

```python
#Average rating per user
amazon.groupby("reviews.username")["reviews.rating"].mean()
```

```python
sns.countplot(data=amazon,x="reviews.rating")
plt.title("Rating Distribution")
plt.show()
```

```python
sns.histplot(amazon,x="reviews.rating",bins=5,kde=True)
plt.show()
```

```python
sns.countplot(amazon, x="reviews.doRecommend")
plt.show()
```

```python
sns.boxplot(amazon,x="reviews.doRecommend",y="reviews.rating")
plt.show()
```

```python
amazon["rev_len"]=amazon["reviews.text"].astype(str).str.len()
sns.histplot(amazon,x="rev_len",bins=20,kde=True)
plt.title("Review length DIstribution")
plt.show()
```

```python
top_brands = amazon["brand"].value_counts().head(10)

sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title("Top 10 Brands by Number of Reviews")
plt.show()
```

```python
top_users = amazon["reviews.username"].value_counts().head(10)
sns.barplot(x=top_users.values, y=top_users.index)
plt.title("Top 10 Reviewers")
plt.show()
```

```python
sns.heatmap(
    amazon[["reviews.rating","reviews.numHelpful","price_numeric"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
