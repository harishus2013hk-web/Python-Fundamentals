rock paper scissor game


```python
import random
game=['Rock', 'Paper', 'Scissor']
round=0
computer_score=0
user_score=0
for i in range(1,4):
    computer=random.choice(game)
    user=int(input("Choose 0-Rock, 1-Paper, 2-Scissor"))
    round+=1
    if user<0 or user>2:
        print("Invalid Choice")
    else:
        u_choice=game[user]
        print(f"Round: {round}\nYou choose:{u_choice}\ncomputer choose:{computer}")

        if u_choice==computer:
            print("It's Draw!")
        elif (u_choice=="Rock" and computer=="Scissor") or \
            (u_choice=="Paper" and computer=="Rock") or \
            (u_choice=="Scissor" and computer=="Paper"):
            print("You Win!\n")
            user_score+=1
        else:
            print("Computer Wins!")
            computer_score+=1
        print(f"Your Score:{user_score}\ncomputer score: {computer_score}\n")
print(f"\nFinal Score\nUser:{user_score}\nComputer: {computer_score}")
if user_score>computer_score:
    print("\U0001F389You Win!")
elif user_score<computer_score:
    print("Computer Wins!")
else:
    print("Draw")
```

Random password generator

```python
import random
import string
letters=list(string.ascii_lowercase)
letter_upper=list(string.ascii_uppercase)
number=list(string.digits)
special=list(string.punctuation)
password=""
print("Welcome to the password generator")
s=int(input("Enter how many small letters you want"))
l=int(input("Enter how many capital letters you want"))
n=int(input("Enter how many numbers you want"))
c=int(input("Enter how many special char you want"))
for i in range(1,s+1):
    password+=random.choice(letters)
for j in range(1,l+1):
    password+=random.choice(letter_upper)
for k in range(1,n+1):
    password+=str(random.choice(number))
for m in range(1,c+1):
    password+=random.choice(special)
password_list=list(password)
random.shuffle(password_list)
password="".join(password_list)
password
```

Hangman Game (correct word guessing game)

```python
import random
lives=6
word_list=['aardvark','baboon','camel']
choose=random.choice(word_list)
print(f"You have to guess the {len(choose)*'_ '} word")
correct_letter=[]
game_over=False
while not game_over:
    guess=input("Guesss the word").lower()
    display=''
    if guess in correct_letter:
        print(f"You have already guessed {guess}")
        continue
    for letter in choose:
        if letter== guess:
            display+=letter
            correct_letter.append(letter)
        elif letter in correct_letter:
            display+=letter
        else:
            display+='_ '
    print(display)
    if guess not in choose:
        lives-=1
        print(f"you have {lives} lives left")
        if lives==0:
            game_over=True
            print("You Lose!")
    if "_" not in display:
        game_over=True
        print("You Win!")
```

Love Calculator

```python
# Love Calculator
def calculate_love_score(name1,name2):
    t='TRUE'
    l='LOVE'
    total_t=0
    total_l=0
    for ch in t:
        count=name1.count(ch.lower()) + name2.count(ch.lower())
        print(f"{ch} occurs {count} times")
        total_t+=count
        
    for ch in l:
        count=name1.count(ch.lower()) + name2.count(ch.lower())
        print(f"{ch} occurs {count} times")
        total_l+=count
    love_score=int(str(total_t) + str(total_l))
    print("Love Score",love_score)
calculate_love_score('Kayne West','Kim Kardishian')
```

Encryption and Decryption of code

```python

alphabet='abcdefghijklmnopqrstuvwxyz'
shift=int(input("Enter the number to shift your character"))
text=input("Entet youe text to encrypt").lower()
encrypt_msg=''
decrypt_msg=''
for i in text:
    pos=alphabet.index(i)+shift
    if pos>=26:
        pos=pos-26
    encrypt_msg+=alphabet[pos]

for j in encrypt_msg:
    pos=alphabet.index(j)-shift
    if pos>=26:
        pos=pos+26
    decrypt_msg+=alphabet[pos]
print(f"Encrypted message={encrypt_msg}")
print(f"Decrypted message={decrypt_msg}")
    

```

Multiple choice quiz

```python
quiz = [
    {
        "question": "What is the capital of France?",
        "options": ["A. Berlin", "B. Madrid", "C. Paris", "D. Rome"],
        "answer": "C"
    },
    {
        "question": "Which language is used for Machine Learning?",
        "options": ["A. Python", "B. HTML", "C. CSS", "D. SQL"],
        "answer": "A"
    },
    {
        "question": "What is 5 + 7?",
        "options": ["A. 10", "B. 12", "C. 14", "D. 15"],
        "answer": "B"
    }
]

score = 0

for q in quiz:
    print("\n" + q["question"])
    
    for option in q["options"]:
        print(option)

    user_answer = input("Enter your answer (A/B/C/D): ").upper()

    if user_answer == q["answer"]:
        print("Correct!")
        score += 1
    else:
        print("Wrong! Correct answer is:", q["answer"])

print(f"\nFinal Score: {score}/{len(quiz)}")
```

To-Do-List (Basic Task Manager)

```python
tasks = []

def todo_app():
    print("\n1. Add Task\n2. View Tasks\n3. Remove Task\n4. Exit")
    while True:
        choice = input("Choose option: ")

        if choice == '1':
            task = input("Enter task: ")
            tasks.append(task)
        elif choice == '2':
            print("Tasks:", tasks)
        elif choice == '3':
            task = input("Enter task to remove: ")
            if task in tasks:
                tasks.remove(task)
            else:
                print("Task not found")
        elif choice == '4':
            break
        else:
            print("Invalid choice")
    print(tasks)
todo_app()
```

Alarm Clock

```python
import time
import winsound

def alarm(set_time):
    while True:
        current_time = time.strftime("%H:%M")
        if current_time == set_time:
            print("Wake up!")
            winsound.Beep(1000, 2000)
            break

alarm("22:57")
```

```python
import random

words = ["python", "data", "science", "machine"]
word = random.choice(words)
guessed = ["_"] * len(word)
lives = 6   # total chances

while "_" in guessed and lives > 0:
    print("\nWord:", " ".join(guessed))
    print(f"Lives left: {lives}")
    letter = input("Guess a letter: ")

    if letter in word:
        for i, l in enumerate(word):
            if l == letter:
                guessed[i] = letter
        print("Good guess!")
    else:
        lives -= 1
        print("Wrong guess!")

if "_" not in guessed:
    print("\n🎉 You guessed it:", word)
else:
    print("\n💀 Game Over! The word was:", word)
```
