Countplot

```python
#Smoker vs Non smoker count (Total)
plt.figure(figsize=(4,3))
a=sns.countplot(data=tips,x="smoker")
plt.title("Smoker count")
plt.bar_label(a.containers[0])
plt.show()
```

```python
#Smoker count on the basis of days
a=sns.countplot(data=tips,x="smoker",hue="day")
plt.title("Day wise smoker count")
for container in a.containers:
    plt.bar_label(container)
plt.show()
```

```python
#Histogram plot showing the total bill variation based on gender
a=sns.histplot(data=tips,x="total_bill",bins=20,kde=True,hue="time",multiple="stack")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()
```

```python
a=sns.histplot(data=tips,x="tip",bins=10,kde=True,hue="sex",multiple="stack")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex", row="time", hue="day")
g.map(sns.scatterplot, "total_bill", "tip").add_legend()
plt.show()
```

```python
time_sum=tips.groupby("time")["total_bill"].sum()
labels=[f"{i}\n{v}" for i,v in zip(time_sum.index,time_sum.values)]
plt.pie(x=time_sum,labels=labels,autopct="%1.1f%%")
plt.title("Lunch Dinner contribution")
plt.show()
```
