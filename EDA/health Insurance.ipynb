```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=insurance, x="age", y="claim", hue="smoker")
plt.show()

```

```python
sns.boxplot(data=insurance, x="diabetes", y="claim")
plt.show()
```

```python

sns.barplot(data=insurance, x="smoker", y="claim")
plt.show()
```

```python
sns.kdeplot(data=insurance, x="bmi", fill=True)
plt.title("BMI Density Plot")
plt.show()
```

```python
sns.violinplot(data=insurance, x="smoker", y="claim",hue="sex")
plt.title("Claim Distribution by Smoker")
plt.show()
```

```python
sns.lineplot(data=insurance, x="age", y="claim")
plt.title("Claim Trend by Age")
plt.show()
```

```python
g = sns.FacetGrid(insurance, col="smoker")
g.map(sns.histplot, "claim")
plt.show()
```

```python
g = sns.FacetGrid(insurance, col="sex", row="smoker")
g.map(sns.scatterplot, "age", "claim")
plt.show()
```

```python
sns.displot(insurance,x="bloodpressure",col="sex",kde=True)
```

```python
corr=insurance.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

```python
sns.regplot(data=insurance,x="weight",y="bloodpressure",ci=95)
plt.show()
```

```python
sns.pairplot(insurance[["age","bmi","claim","smoker"]],hue="smoker")
```

```python
sns.lmplot(data=insurance,x="weight",y="age")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=insurance,x="hereditary_diseases")
plt.show()
```
