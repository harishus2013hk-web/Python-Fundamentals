```python

sale["Order Date"]=pd.to_datetime(sale["Order Date"],dayfirst=True)
sale["Ship Date"]=pd.to_datetime(sale["Ship Date"],dayfirst=True)
```

```python
pivot=pd.pivot_table(data=sale,values="Sales",index=["Category","Sub-Category","Segment"],aggfunc="sum")

#category totals
cat_total=sale.groupby("Category")["Sales"].sum().to_frame("Sales")
cat_total["Sub-Category"]="Category Total"
cat_total["Segment"]=""
cat_total.set_index(["Sub-Category","Segment"],append=True,inplace=True)

#Sub-Category totals
sub_cat_total=sale.groupby(["Category","Sub-Category"])["Sales"].sum().to_frame("Sales")
sub_cat_total["Segment"]="Sub-Category Total"
sub_cat_total.set_index(["Segment"],append=True,inplace=True)

#Segment wise total
seg_total=sale.groupby("Segment")["Sales"].sum().to_frame("Sales")
seg_total["Category"]="Segment Total"
seg_total["Sub-Category"]=""
seg_total.set_index(["Category","Sub-Category"],append=True,inplace=True)
seg_total=seg_total.reorder_levels(["Category","Sub-Category","Segment"])

#Combine everything
final_pivot=pd.concat([pivot,cat_total,sub_cat_total,seg_total]).sort_index()
print(final_pivot)
```

```python
sale["Order_Year"]=sale["Order Date"].dt.year
yearly_sales=sale.groupby("Order_Year")["Sales"].sum().reset_index()
print(yearly_sales)
sns.lineplot(data=yearly_sales,x="Order_Year",y="Sales")
```

```python
plt.figure(figsize=(12,4))
top_state=sale.groupby("State")["Sales"].sum().nlargest(10).reset_index()
sns.barplot(data=top_state,x="State",y="Sales")
plt.show()
```

```python
pie_plot=sale.groupby("Region")["Sales"].sum()
label=[f"{i}\n{j}" for i,j in zip(pie_plot.index,pie_plot.values)]
plt.pie(pie_plot,labels=label,autopct="%1.1f%%")
plt.title("Region wise sales contribution")
plt.show()
```
