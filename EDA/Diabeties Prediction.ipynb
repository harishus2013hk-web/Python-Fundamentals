```python
pd.pivot_table(
    data=sugar,
    values=["bmi", "HbA1c_level"],
    index=["gender", "smoking_history"],
    columns="diabetes",
    aggfunc=["mean", "median", "min","max"]
)
```

```python
sns.displot(data=sugar,x="bmi",col="gender",bins=30,kde=True)
plt.show()
```

```python
sns.displot(data=sugar,x="bmi",hue="diabetes",kde=True,multiple="stack",bins=30)
plt.show()
```

```python
sns.displot(data=sugar,x="HbA1c_level",hue="gender",kde=True,bins=30,multiple="stack")
plt.show()
```

```python
sns.displot(data=sugar,x="blood_glucose_level",col="smoking_history",kde=True)
plt.show()
```

```python
g=sns.FacetGrid(data=sugar,col="gender",row="diabetes")
g.map(sns.scatterplot,"age","bmi")
g.add_legend()
plt.show()
```

```python
g=sns.FacetGrid(data=sugar,col="smoking_history",col_wrap=3)
g.map(sns.histplot,"age",kde=True)
plt.show()
```

```python
sns.countplot(data=sugar,x="diabetes",hue="gender")
plt.title("Diabetes count by Gender")
plt.show()
```

```python
sns.countplot(data=sugar,x="smoking_history",hue="gender")
plt.title("Smoking History count by Gender")
plt.show()
```

```python
grp_bmi=sugar.groupby("gender")["bmi"].mean().reset_index()
a=sns.barplot(data=grp_bmi,x="gender",y="bmi")
plt.title("Average BMI by Gender")
a.bar_label(a.containers[0])
plt.show()
```

```python
grp_glucose=sugar.groupby(["gender","smoking_history"])["blood_glucose_level"].mean().reset_index()
sns.barplot(data=grp_glucose,x="smoking_history",y="blood_glucose_level",hue="gender")
plt.show()
```

```python
pivot_hba1c = pd.pivot_table(
    data=sugar,
    values="HbA1c_level",
    index="smoking_history",
    columns="diabetes",
    aggfunc="mean"
)

pivot_hba1c.plot(kind="bar")
plt.title("Mean HbA1c by Smoking History & Diabetes")
plt.xticks(rotation=45)
plt.show()
```

```python
sugar["age_group"]=pd.cut(sugar["age"],bins=[0,30,45,60,100],labels=["Young","Adult","Middle","Senior"])
pivot_age=pd.pivot_table(data=sugar,values="bmi",index="age_group",columns="diabetes",aggfunc="mean").reset_index()
pivot_age.plot(x="age_group",marker=">")
plt.show()
pivot_age
```

```python
sns.lineplot(data=sugar,x="age",y="bmi",ci=None)
plt.show()
```

```python
sns.boxplot(data=sugar, x="diabetes", y="HbA1c_level")
plt.show()
```

```python
sns.violinplot(data=sugar, x="diabetes", y="HbA1c_level")
plt.show()
```

```python
sns.pairplot(sugar[["bmi","HbA1c_level","blood_glucose_level"]])
plt.show()
```

```python
age_group_sugar=sugar.groupby("age_group")["diabetes"].value_counts().reset_index()
age_group_sugar
a=sns.barplot(data=age_group_sugar,x="age_group",y="count",hue="diabetes")
a.bar_label(a.containers[0])
a.bar_label(a.containers[1])
plt.show()
```

```python
sugar.describe()
col=["age","bmi","HbA1c_level","blood_glucose_level"]
fig, axs = plt.subplots(len(col), 1, figsize=(5, 14), dpi=95)
for i, c in enumerate(col):
    axs[i].boxplot(sugar[c],vert=False)
    axs[i].set_ylabel(c)
plt.tight_layout()
plt.show()
```
