```python
pd.pivot_table(data=cancer,values=["BMI","PhysicalActivity","AlcoholIntake"],
               index=["Gender","Smoking","Diagnosis","CancerHistory"],
               aggfunc=["mean","min","max","sum"])
```

```python
cancer["Gender_label"]=cancer["Gender"].map({1:"Male",0:"Female"}).fillna("Other")
```

```python
cancer["CancerHistory"]=cancer["CancerHistory"].map({0:"No",1:"Yes"})
cancer["Diagnosis"]=cancer["Diagnosis"].map({0:"No",1:"Yes"})
cancer["Smoking"]=cancer["Smoking"].map({0:"No",1:"Yes"})
```

```python
cancer["Age_Group"]=pd.cut(cancer["Age"],bins=[20,30,60,100],labels=["Young","Adult","Senior"])
cancer.head()
```

```python
num_cols = ["Age", "BMI", "PhysicalActivity", "AlcoholIntake"]

for col in num_cols:
    plt.figure(figsize=(6,2))
    sns.histplot(data=cancer, x=col,bins=20, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in num_cols:
    sns.boxplot(data=cancer, x="Age_Group",y=col)
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
cat_cols = [
    "Gender_label", "Smoking", "GeneticRisk",
    "CancerHistory", "Diagnosis", "Age_Group"
]

for col in cat_cols:
    plt.figure(figsize=(5,3))
    a=sns.countplot(data=cancer, x=col)
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.bar_label(a.containers[0])
    plt.show()
```

```python
sns.boxplot(data=cancer, x="Diagnosis", y="Age")
plt.title("Age vs Diagnosis")
plt.show()
```

```python
sns.violinplot(data=cancer, x="Diagnosis", y="BMI")
plt.title("BMI Distribution by Diagnosis")
plt.show()
```

```python
sns.countplot(data=cancer, x="Gender_label", hue="Diagnosis")
plt.title("Gender vs Diagnosis")
plt.show()
```

```python
sns.countplot(data=cancer, x="Smoking", hue="CancerHistory")
plt.title("Smoking vs Cancer History")
plt.show()
```

```python
sns.pairplot(
    cancer,
    vars=["BMI", "PhysicalActivity", "AlcoholIntake"],
    hue="Diagnosis"
)
plt.show()
```

```python
corr = cancer[
    ["Age", "BMI", "PhysicalActivity", "AlcoholIntake", "Gender"]
].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.stripplot(data=cancer, x="Diagnosis", y="Age", jitter=True)
plt.title("Age vs Diagnosis (Strip Plot)")
plt.show()
```

```python
sns.swarmplot(data=cancer, x="Diagnosis", y="BMI")
plt.title("BMI vs Diagnosis (Swarm Plot)")
plt.show()
```

```python
sns.kdeplot(data=cancer, x="Age", hue="Diagnosis", fill=True)
plt.title("Age Density by Diagnosis")
plt.show()
```

```python
sns.displot(data=cancer,x="PhysicalActivity",col="Age_Group",kind="kde")
plt.show()
```

```python
sns.lineplot(data=cancer,x="Age",y="PhysicalActivity")
plt.show()
```
