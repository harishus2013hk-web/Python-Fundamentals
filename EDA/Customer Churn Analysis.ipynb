```python
# Churn Distribution
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
# Histogram for tenure
sns.histplot(df["Tenure"], kde=True)
plt.title("Distribution of Tenure")
plt.show()
```

```python
# Histogram for Spends
sns.histplot(df["Total Spend"], kde=True)
plt.title("Distribution of Spends")
plt.show()
```

```python
# Countplot for Contract type
sns.countplot(x="Contract Length", data=df)
plt.title("Contract Type Distribution")
plt.show()
```

```python
# Boxplot: Age vs Churn
sns.boxplot(x="Churn", y="Age", data=df)
plt.title("Age vs Churn")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr=df.corr(numeric_only=True)
sns.heatmap(data=corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Age distribution split by Churn
sns.displot(data=df, x="Age", hue="Churn", kde=True, bins=20,multiple="stack")
plt.title("Age Distribution by Churn")
plt.show()
```

```python
# Usage Frequency vs Total Spend
sns.scatterplot(x="Usage Frequency", y="Total Spend", data=df, hue="Churn", style="Churn")
plt.title("Usage Frequency vs Total Spend (colored by Churn)")
plt.show()
```

```python
# Pairplot for selected numerical features
sns.pairplot(df[["Age", "Tenure", "Usage Frequency", "Total Spend", "Churn"]], hue="Churn")
plt.show()
```
