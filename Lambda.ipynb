```python
add=lambda a,b:a+b
print(add(5,10))
```

```python
num=[1,2,3,4,5,6,7,8]
even=list(filter(lambda x:x%2==0,num))
print(even)
```

```python
name=["abc","abcde","ab","abcd","a"]
name.sort(key=lambda x:len(x))
print(name)
```

```python
name=["abc","abcde","ab","abcd","a"]
name.sort(key=lambda x:len(x),reverse=True)
print(name)
```

```python
#Check Even or Odd
even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(even(7))   # Output: Odd
```

```python
#Square of a Number
square = lambda x: x ** 2
print(square(6))   # Output: 36
```

```python
#Sort List of Tuples by Second Value
data = [(1, 5), (2, 3), (4, 1)]
data.sort(key=lambda x: x[1])
print(data)   # Output: [(4, 1), (2, 3), (1, 5)]
```

```python
#Find Maximum of Two Numbers
maximum = lambda a, b: a if a > b else b
print(maximum(10, 25))   # Output: 25
```

```python
#Reverse a String
reverse = lambda s: s[::-1]
print(reverse("Python"))   # Output: nohtyP
```
