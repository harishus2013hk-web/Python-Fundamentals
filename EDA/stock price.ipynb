```python
stock["month"]=stock["Date"].dt.month
```

```python
stock["month_name"]=stock["Date"].dt.month_name()
```

```python
stock["Day"]=stock["Date"].dt.day_name()
```

```python
stock.head()
```

```python
stock.groupby(["month_name","Day"])[["Stock_1","Stock_2","Stock_3","Stock_4","Stock_5"]].agg(["min","max","mean"])
```

```python
month_order=["January","February","March","April","May","June",
               "July","August","September","October","November","December"]
days_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

stock["month_name"]=pd.Categorical(stock["month_name"],categories=month_order,ordered=True)

stock["Day"]=pd.Categorical(stock["Day"],categories=days_order,ordered=True)

result=stock.groupby(
    ["month_name","Day"])[["Stock_1","Stock_2","Stock_3","Stock_4","Stock_5"]].agg(["min","max","mean"]).sort_index(level=["month_name","Day"])

print(result)
```

```python
col=["Stock_1","Stock_2","Stock_3","Stock_4","Stock_5"]
for i in col:
    sns.lineplot(data=stock,x="month_name",y=i,label=i)
plt.title("Stock trend over month")
plt.xticks(rotation=45)
plt.show()
```

```python
stock_melt=stock.melt(id_vars="month",value_vars=["Stock_1","Stock_2","Stock_3","Stock_4","Stock_5"],var_name="Stock",value_name="Price")
```

```python
sns.relplot(data=stock_melt,x="month",y="Price",col="Stock",kind="line",col_wrap=3)
plt.show()
```

```python
sns.histplot(data=stock_melt,x="Price",hue="Stock",bins=10,kde=True,multiple="stack")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=stock_melt, x="Stock", y="Price")
plt.title("Stock Price Spread")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=stock_melt, x="Stock", y="Price")
plt.title("Stock Price Density")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=stock_melt, x="Stock", y="Price", estimator="mean")
plt.title("Average Stock Price")
plt.show()
```

```python
corr = stock[["Stock_1","Stock_2","Stock_3","Stock_4","Stock_5"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Stock Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(
    data=stock,
    x="Stock_1",
    y="Stock_3"
)
plt.title("Stock 1 vs Stock 3")
plt.show()
```

```python
sns.pairplot(
    stock[["Stock_1","Stock_2","Stock_3","Stock_4","Stock_5"]]
)
plt.show()
```
