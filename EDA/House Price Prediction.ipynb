```python
hp["price_category"]=pd.cut(hp["Price"],bins=[0,200000,500000,1000000],labels=["Low","Medium","High"])
hp["Year_Group"]=pd.cut(hp["YearBuilt"],bins=[1900,1925,1950,1975,2000,2025],labels=["1900-1925","1926-1950","1951-1975","1976-2000","2001-2025"])
```

```python
pd.pivot_table(data=hp,values="Price",index="Area",columns="Bedrooms")
```

```python
sns.lineplot(data=hp,x="Bedrooms",y="Price",hue="Location",ci=None)
```

```python
a=sns.barplot(data=hp,y="Price",x="Condition",hue="Location")
for container in a.containers:
    a.bar_label(container,rotation=90,label_type='center')
```

```python
sns.pairplot(data=hp,vars=["Price","Area","Bedrooms","Floors"],hue="price_category")
plt.show()
```

```python
sns.lineplot(data=hp,x="Year_Group",y="Price",errorbar=None)
plt.show()
```

```python
sns.histplot(data=hp,x="Price",kde=True,bins=50)
plt.xlabel("Price in Millions")
plt.show()
```

```python
sns.displot(data=hp,x="Price",col="Bedrooms",kde=True,hue="Garage",multiple="stack")
plt.show()
```

```python
sns.scatterplot(
    data=hp,
    x="Area",
    y="Price",
    hue="price_category"
)
plt.title("Area vs Price by Category")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = hp[['Area', 'Bedrooms', 'Bathrooms', 'Floors', 'Price']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.lineplot(data=hp, x="YearBuilt", y="Price",ci=None)
plt.title("YearBuilt vs Price")
plt.show()
```

```python
sns.violinplot(data=hp, x="Bedrooms", y="Price")
plt.title("Price Distribution by Bedrooms")
plt.show()
```

```python
g=sns.FacetGrid(data=hp,col="Location")
g.map(sns.histplot,"Price",bins=30,kde=True)
plt.show()
```

```python
g=sns.FacetGrid(data=hp,col="Condition",hue="Garage",col_wrap=4,height=4)
g.map(sns.scatterplot,"Area","Price")
g.add_legend()
```

```python
g=sns.FacetGrid(data=hp,col="Year_Group")
g.map(sns.kdeplot,"Price")
```
